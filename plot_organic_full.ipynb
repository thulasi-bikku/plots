```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
col_names = ['shelf', 'book', 'page', 'wl', 'n', 'k']
```

```python
df = pd.read_excel('Organic_full.xlsx', names = col_names)
```

```python
df.shape
```

```python
df.sample()
```

```python
# list of the books
x = df.book.value_counts().keys()
booklst = list()
for i in x:
    booklst.append(i.capitalize())
# Total number of counts of each book, respectively    
bc = df.book.value_counts().values
book_count = list()
for i in bc:
    book_count.append(i)
```

```python
for i in range(len(booklst)):
    print(booklst[i], ':', book_count[i])
```

```python
df.to_csv('components.csv', index = False)
df
```
